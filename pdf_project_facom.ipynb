```python

# csv name
csv_entered_name = 'facom_url'
csv_entered_name = csv_entered_name + '.csv'
csv_outgiven_name = 'outgiven_' + csv_entered_name

# pdf_entered_nema
pdf_entered_name = 'facom'
pdf_entered_name = pdf_entered_name + '.pdf'
pdf_outgiven_name = 'outgiven_' + pdf_entered_name

starting_page = 30
ending_page = 40

import pymupdf  # PyMuPDF library for handling PDF files
import pandas as pd  # Pandas library for data manipulation

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_entered_name, sep=";")

# Convert the 'sku' column to uppercase and clean any non-breaking spaces
df['sku'] = df['sku'].apply(lambda x: str(x).replace(u'\xa0', u'').upper())

# Print the entire DataFrame for inspection
print(df.to_string())

# Convert the DataFrame to a list of lists (each row becomes a list)
df_list = df.values.tolist()

# Print specific rows from the list for inspection
print("Row 252:", df_list[252])
print("Row 243:", df_list[243])

# Initialize lists to store results
entered_list = []
lista = df_list

# Open the PDF document
doc = pymupdf.open(pdf_entered_name)

# Initialize lists to store found SKUs and their counts
olokliro = []
nropages = doc.page_count

# Loop through all pages in the PDF
for page in doc.pages(starting_page, ending_page, 1):
    # Extract text from the current page
    s = page.get_text()
    
    # Check each SKU in the list
    for i in lista:
        if i[0] in s:
            olokliro.append(i[0])

# Print the number of SKUs found in the PDF
print("Total SKUs found:", len(olokliro))

# Remove duplicate SKUs
non_dup_list = list(set(olokliro))

# Print the number of unique SKUs
print("Total unique SKUs found:", len(non_dup_list))

# Filter the original list to only include SKUs that were found
lista_new = [i for i in lista if i[0] in non_dup_list]

# Print the filtered list and counts
print("Filtered SKU list:", lista_new)
print("Unique SKUs list:", non_dup_list)
print("Filtered list length:", len(lista_new))
print("Unique SKUs length:", len(non_dup_list))



# Open the PDF document again for annotation
doc = pymupdf.open(pdf_entered_name)

# Initialize list to store SKUs that have been annotated
entered_list = []
nropages = doc.page_count
print("Number of pages in PDF:", nropages)

# Loop through a limited number of pages (change as needed)
for page in doc.pages(starting_page, ending_page , 1): 
    print('Processing page:', page.number)
    nri = 0
    
    # Loop through the filtered SKU list for annotation
    for i in lista_new:
        nri += 1
        print('Processing SKU number:', nri)
        
        # Search for each SKU in the current page
        for rect in page.search_for(i[0]):
            # Add SKU to the entered list
            entered_list.append(i)
            print('Found SKU rectangle:', rect)
            
            # Add highlight annotation to the found SKU text
            page.add_highlight_annot(rect)
            
            # Insert a clickable link to the SKU URL
            page.insert_link({
                "kind": pymupdf.LINK_URI,
                "from": rect,
                "to": i[1],
                "page": -1,
                "uri": i[1]
            })
            print('Added URL link for SKU')

# Print confirmation of loop completion
print('Completed annotation loop')

# Convert the annotated SKUs list to a DataFrame and save as CSV
df_entered = pd.DataFrame(entered_list)
df_entered.to_csv(csv_outgiven_name, header=False)
print('CSV file saved as csv_name_whole.csv')

# Save the modified PDF with annotations
doc.save(pdf_outgiven_name, garbage=2, deflate=True)
print('PDF file saved as pdf_name_whole.pdf')

# Close the PDF document
doc.close()

```
