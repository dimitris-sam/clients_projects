```python
# cell 1
# extracts the images in png format, saves the spareparts 
# saves the extracted images names in images_names_list
import fitz  # PyMuPDF
import pandas as pd

pdf_path = "HR5212.pdf"  # Replace with your PDF file path

# Open the PDF
doc = fitz.open(pdf_path)

# Initialize an empty list to collect the table rows
all_extracts = []
image_counter = 1
images_names_list = []

# Loop through each page in the PDF
for page_num in range(len(doc)):
    page = doc.load_page(page_num)  # Load the actual page
    
    # Extract the images on this page (moved before table processing)
    image_list = page.get_images(full=True)
    
    # Save the images
    for img_index, img in enumerate(image_list):
        xref = img[0]  # XREF of the image
        # Get the image data
        image = doc.extract_image(xref)
        image_bytes = image["image"]
        # Define the filename for the extracted image
        image_filename = f"page_image_{image_counter}.png"
        with open(image_filename, "wb") as img_file:
            img_file.write(image_bytes)
            print(f"Saved image: {image_filename}")
            images_names_list.append(image_filename)
        image_counter += 1
    
    # Now process tables
    tabs = page.find_tables()  # Find tables on the page
    if not tabs.tables:  # If no tables are found on the page, skip table processing
        continue

    # Access the first table (assuming you are only interested in the first table)
    tab = tabs[0]
    extract = tab.extract()  # Get text for all table cells
    all_extracts.extend(extract)  # Append the extracted rows to the list

    # Extract the images on this page
    image_list = page.get_images(full=True)

    # Save the images
    for img_index, img in enumerate(image_list):
        xref = img[0]  # XREF of the image

        # Get the image data
        image = doc.extract_image(xref)
        image_bytes = image["image"]

        # Define the filename for the extracted image
        image_filename = f"page_image_{image_counter}.png"
        with open(image_filename, "wb") as img_file:
            img_file.write(image_bytes)
            print(f"Saved image: {image_filename}")
            images_names_list.append(image_filename)

        image_counter += 1

# If there are any extracts, assume the first row is the header
if all_extracts:
    column_names = list(all_extracts[0])  # Assuming the first row is the header
    all_extracts = all_extracts[1:]  # Remove the header from the row data
else:
    column_names = None

# Create DataFrame from the table data
df = pd.DataFrame(all_extracts, columns=column_names)
# Remove empty rows (rows where all values are NaN)
df = df.dropna(how='all')
# Adjust pandas settings to display all rows (optional)
pd.set_option('display.max_rows', None)  # Show all rows

# Display the DataFrame
print(df)

# Optionally, save the DataFrame to an Excel file
df.to_excel("extracted_table.xlsx", index=False)



# The string you're searching for in the second column
search_string = '631937-6'

# Iterate through the DataFrame and check each row
for index, row in df.iterrows():
    # Use the second column header dynamically
    if row[df.columns[1]] == search_string:
        print(row[df.columns[0]])  # Print value from the first column
        break


# βγαζει τις εικονες και βγαζει και σε df τον πινακα
print(images_names_list)
```

```python
import fitz  # PyMuPDF
from PIL import Image
import pymupdf

# Input file paths
# original_pdf_path = "HR2631F.pdf"

output_pdf_path_list = []
for position_2, image_name in enumerate(images_names_list):
    image_path = image_name


    # Output file paths
    output_pdf_path = f"annotated_pdf_{position_2}.pdf"

    # Step 1: Get page dimensions from original PDF
    doc = fitz.open(pdf_path)
    page = doc[0]
    page_width_points = page.rect.width
    page_height_points = page.rect.height
    doc.close()

    # Step 2: Open the image to get its pixel dimensions
    image = Image.open(image_path)
    image_width_pixels, image_height_pixels = image.size

    # Step 3: Create a new PDF with the image
    pdf = fitz.open()
    pdf.new_page(width=page_width_points, height=page_height_points)
    image_rect = fitz.Rect(0, 0, page_width_points, page_height_points)
    # pdf[position].insert_image(image_rect, filename=image_path)
    pdf[0].insert_image(image_rect, filename=image_path)

    # Step 4: Add clickable links
    for entry in new_dict_list:
        if entry.get('position') == position_2:  # Only process if position is 0
            x, y, w, h = entry['location']
            print(entry['url_key'])
            print(type(entry['url_key']))    
            # Transform image pixels to PDF points
            x_pt = x / image_width_pixels * page_width_points
            y_pt = y / image_height_pixels * page_height_points
            w_pt = w / image_width_pixels * page_width_points
            h_pt = h / image_height_pixels * page_height_points
            
            rect = fitz.Rect(x_pt, y_pt, x_pt + w_pt, y_pt + h_pt)
            pdf[0].add_highlight_annot(rect)
            
            # Ensure the URL has a valid scheme
            url = entry['url_key']
            if not url.startswith(('http://', 'https://')):
                url = 'http://' + url

            # Insert the clickable web link
            pdf[0].insert_link({
                "kind": pymupdf.LINK_URI,
                "from": rect,
                "uri": url
            })


    # Save single-page PDF and track path
    pdf.save(output_pdf_path)
    pdf.close()
    output_pdf_path_list.append(output_pdf_path)

print(f"Saved individual PDFs: {output_pdf_path_list}")

# Step 5: Merge all individual PDFs into one
merged_pdf = fitz.open()
for pdf_path in output_pdf_path_list:
    single_pdf = fitz.open(pdf_path)
    merged_pdf.insert_pdf(single_pdf)
    single_pdf.close()

merged_pdf.save("merged_annotated.pdf")
merged_pdf.close()
print("Merged PDF saved as merged_annotated.pdf")

# # Save the final PDF
# pdf.save(output_pdf_path)
# pdf.close()

# print(f"Annotated PDF saved to {output_pdf_path}")

print(output_pdf_path_list)
```
